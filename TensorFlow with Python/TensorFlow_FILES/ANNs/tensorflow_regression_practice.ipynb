```python
X=df.drop('price',axis=1)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train.values,validation_data=(X_test,y_test.values),batch_size=128,epochs=400)
# validation_data: Data on which to evaluate the loss and any model metrics at the end of each epoch. The model will not be trained on this data.
#validation_data measure how well our model is performing during the training process without effecting the trained result
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
pd.DataFrame(model.history.history)
```

```python
pd.DataFrame(model.history.history).plot() # we can tell that there's no overfitting in our model
```

```python
pred=model.predict(X_test)
```

```python
mean_absolute_error(pred,y_test)
```

```python
mean_squared_error(pred,y_test)
```

```python
np.sqrt(mean_squared_error(pred,y_test))
```

```python
explained_variance_score(y_test,pred)
```
