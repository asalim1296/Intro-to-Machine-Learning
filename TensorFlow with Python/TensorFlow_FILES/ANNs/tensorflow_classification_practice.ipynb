```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('benign_0__mal_1',axis=1).values # values to return a numpy array 
```

```python
y=df['benign_0__mal_1'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model=Sequential()
```

```python
model.add(Dense(30,activation='relu'))

model.add(Dense(15,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy') 
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=600,verbose=1)
```

```python
pd.DataFrame(model.history.history).plot() # example of overfitting; model performed better on training data(loss)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

## we use early stopping to control and monitor val_loss/ loss from increasing so we prevent overfitting in case we set a high number of epochs

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
""" In `min` mode, training will stop when the quantity monitored has stopped decreasing; 
in `"max"` mode it will stop when the quantity monitored has stopped increasing;"""
```

### verbose=1 will show you an animated progress bar like this:
